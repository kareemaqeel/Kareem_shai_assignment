```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/sample_data/Salaries.csv')
data.head()
```

```python
data.info()
```

```python
# Check for missing values in each column
missing_values = data.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
null_cloms = data.columns[data.isna( ).any( )]
for i in null_cloms :
  value = data[i].mean( ) if data[i].dtype != "object" else data[i].mode( )
  data[i].fillna(value , inplace= True)
```

```python
data['TotalPay'].agg(['mean', 'median', 'max', 'min'])
```

```python
data['TotalPay'].std()
```

```python
TotalPay_ = 'TotalPay'

TotalPay_data = data[TotalPay_]


TotalPay_counts = TotalPay_data.value_counts()


sorted_TotalPay = TotalPay_counts.sort_index()


plt.bar(sorted_TotalPay.index, sorted_TotalPay.values)

plt.xlabel(TotalPay_)
plt.ylabel('Frequency')
plt.title(f'Distribution of {TotalPay_}')
```

```python

JobTitle_name = 'JobTitle'


JobTitle_data = data[JobTitle_name]


value_counts = JobTitle_data.value_counts()


plt.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%')

plt.title(f'Distribution of {JobTitle_name}')
```

```python
data.groupby (['Year', 'JobTitle'])['TotalPay'].mean()
```

```python

TotalPay_ = 'TotalPay'
BasePay_ = 'BasePay'

TotalPay_data = data[TotalPay_]
BasePay_data = data[BasePay_]

plt.scatter(BasePay_data, TotalPay_data)

plt.xlabel( BasePay_)
plt.ylabel(TotalPay_)
plt.title(f'Scatter Plot: {TotalPay_} vs {BasePay_}')
```

```python

TotalPay_ = 'TotalPay'
OtherPay_ = 'OtherPay'

TotalPay_data = data[TotalPay_]
OtherPay_data = data[OtherPay_]

plt.scatter(TotalPay_data, OtherPay_data)

plt.xlabel(TotalPay_)
plt.ylabel(OtherPay_)
plt.title(f'Scatter Plot: {TotalPay_} vs {OtherPay_}')
```

```python

TotalPay_ = 'TotalPay'
OvertimePay_ = 'OvertimePay'

TotalPay_data = data[TotalPay_]
OvertimePay_data = data[OvertimePay_]

plt.scatter(TotalPay_data, OvertimePay_data)

plt.xlabel(TotalPay_)
plt.ylabel(OvertimePay_)
plt.title(f'Scatter Plot: {TotalPay_} vs {OvertimePay_}')
```

It is noted that employees with low and medium salaries increase due to the additional working hours they put into their work, unlike employees with high salaries.
